# Workload Distribution 


`Workload scenarios: Homogeneous and Heterogeneous` 
`n = 10,000 samples`


```python
import pandas as pd
import seaborn as sns
import json

import matplotlib.pyplot as plt

# Read JSON file and create DataFrame
with open('data/homogeneous_short_10000.json', 'r') as f:
    data = json.load(f)
df = pd.DataFrame(data)

# Plot SubmissionTime distribution
plt.figure(figsize=(10, 4))
sns.histplot(df['SubmissionTime'], bins=30, kde=True)
plt.title('Distribution of SubmissionTime')
plt.xlabel('SubmissionTime')
plt.ylabel('Count')
plt.show()

# Plot Walltime distribution
plt.figure(figsize=(10, 4))
sns.histplot(df['Walltime'], bins=30, kde=True)
plt.title('Distribution of Walltime')
plt.xlabel('Walltime')
plt.ylabel('Count')
plt.show()

# Plot JobType distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='JobType', data=df)
plt.title('Distribution of JobType')
plt.xlabel('JobType')
plt.ylabel('Count')
plt.show()

# Plot HPCSite distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='HPCSite', data=df)
plt.title('Distribution of HPCSite')
plt.xlabel('HPCSite')
plt.ylabel('Count')
plt.show()
```
